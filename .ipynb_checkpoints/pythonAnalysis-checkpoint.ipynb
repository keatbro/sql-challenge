```python
# Import Dependencies
import pandas as pd
from sqlalchemy import create_engine, inspect
from sqlalchemy.orm import Session

# Create engine connecting to Postgres DB
engine = create_engine('postgresql://***:***@localhost:5433/sql-challenge')
```

```python
# Store all salary data in a variable
salaries = pd.read_sql('select * from salaries', engine)
```

```python
# Plot a simple histogram to visualize distribution of observed salaries
salaries['salary'].hist(alpha=0.6)
```

```python
# Plot title and average salary by creating a dataframe combining both tables
df = pd.read_sql('select t.title as title, avg(s.salary) as avg_salary from salaries s join titles t on s.emp_no = t.emp_no group by 1 order by 2', engine)
df = df.set_index('title')
df.plot.bar()
```
